# Split the data

```python
X = dataframe.iloc[:, :-1].values
y = dataframe.iloc[:, -1].values
print(X)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_predictions = regressor.predict(X_test)

bestThreshold = 0.0
bestScore = 0

index = 0.00001

while index <= 0.1:
    y_temp = np.where(y_predictions >= index, 1, 0)
    score = r2_score(y_test, y_temp)

    if(score > bestScore):
        bestThreshold = index
    index += 0.00001

y_predictions = np.where(y_predictions >= bestThreshold, 1, 0)
score = r2_score(y_test, y_predictions)
print('bestThreshold:',bestThreshold)
print(score)
```
